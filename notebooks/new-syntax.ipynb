```python
MINMAX = 10
rng = np.random.default_rng(333)

tre = toytree.rtree.bdtree(50, seed=rng.integers(1e9)).mod.node_scale_root_height(5e6)
tre = tre.set_node_values("Ne",
    {i: rng.uniform(5e4, 5e5) for i in range(tre.nnodes)}
)
ncolors = []
for i in tre.idx_dict:
    if tre.idx_dict[i].is_root() or len(tre.idx_dict[i].up) > MINMAX:
        ncolors.append('orange')
    else:
        ncolors.append('darkteal')
#ncolors = ['orange' if len(tre.idx_dict[i]) >= MINMAX else 'darkteal' for i in tre.idx_dict]
nsizes = [20 if len(tre.idx_dict[i]) > 1 else 0 for i in tre.idx_dict]

tre.draw(
    ts='p', 
    node_colors=ncolors, 
    node_sizes=nsizes, 
    layout='d',
    #node_labels=False,
    width=800,
    edge_type='c',
);
```

```python
tree = toytree.rtree.unittree(10, seed=333, treeheight=2e6)
tree = tree.set_node_values("Ne", {i: np.random.uniform(1e3, 1e4) for i in tree.idx_dict})
tree.draw(ts='p', admixture_edges=(3, 0));
```

```python
model = ipcoal.Model(
    tree=tree, 
    nsamples=4, 
    seed_trees=3, 
    seed_mutations=34,
    subst_model="jc69",
)
model.sim_loci(10, 10)
```

```python
model.write_loci_to_hdf5('test', '/tmp', diploid=True)
```

```python
from ipcoal.io.transformer import Transformer
txf = Transformer(model.seqs, model.alpha_ordered_names, model.alleles, diploid=True)
txf.names
```
