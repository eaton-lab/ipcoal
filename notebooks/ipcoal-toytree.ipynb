```python
tre = toytree.rtree.unittree(10, treeheight=1e6)
tre = tre.set_node_values("Ne", default=1e5)
tre.draw(node_labels=("Ne", 1, 1));
```

```python
mod = ipcoal.Model(toytree.rtree.baltree(6, treeheight=1e5), Ne=1e6, nsamples=5)
mod.sim_trees(100)
```

```python
mtre = toytree.mtree(mod.df.genealogy.tolist())
```

```python
mtre.draw(layout='d', shared_axes=True);
```
