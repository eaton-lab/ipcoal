### verify time units

verify that transformed trees have tips that align at zero as expected, including when ipcoal tries to ultametricize trees. Units in float vs. ints, etc.

```python
import toytree
import ipcoal
```

```python
tre = toytree.rtree.unittree(10, treeheight=1e5)
tre = toytree.rtree.bdtree(10).mod.node_scale_root_height(1e5)
tre.draw(ts='o');
print(tre.get_node_coordinates().astype(int))
```

```python
model = ipcoal.Model(tree=tre, nsamples=1)
model.sim_trees()
model.draw_genealogy(0);
```

```python
model.sim_loci(1, 10)
model.draw_seqview();
```
