## VCF results format

```python
import toytree
import ipcoal
```

```python
ipcoal.__version__  # must be 0.1.6 or higher
```

### Define the species tree model

```python
tree = toytree.rtree.unittree(6, 1e6)
```

### Define the coalescent model

```python
model = ipcoal.Model(tree, Ne=1e5, nsamples=2)
```

### Simulate data

```python
model.sim_loci(10, 100)
```

### Show results

```python
model.df.head()
```

```python
model.draw_seqview(0, 0, 30, show_text=True);
```

```python
model.write_vcf(diploid=True).head(10)
```
